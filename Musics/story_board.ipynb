```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.signal import hilbert
```

```python
# write a function that plots music_target_env and music_test_env in the same figure and fades music_target_env
def plot_fade(music_target_env, music_test_env, fs):
    t = np.linspace(0,np.shape(music_test_env)[0]/fs,np.shape(music_test_env)[0])
    print(t)
    plt.figure(figsize=(10,5))
    plt.plot(t,music_test_env,color='red')
    music_target_env = music_target_env[0:np.shape(music_test_env)[0]]
    plt.plot(t,music_target_env,alpha=0.5,color='blue')
    plt.ylim(0,1.1*np.max(music_target_env))
    plt.xlabel('Time (samples)')
    plt.ylabel('Amplitude')
    plt.show()
```

```python
fs,music = wavfile.read('../Musics/Archive/robot_test.wav')
music_test_hil = hilbert(music[:,0])
music_test_env = np.abs(music_test_hil)
fs,music = wavfile.read('../Musics/roboalgo.wav')
music_target_hil = hilbert(music[:,0])
music_target_env = np.abs(music_target_hil)
plot_fade(music_target_env,music_test_env,fs)



```
